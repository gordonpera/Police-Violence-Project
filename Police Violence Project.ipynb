Random Forest - y:Charged

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np

max_depth = range(1,11)

# first we create a  list to store the average RMSE for each value of max_depth
RMSE_scores = []

from sklearn.cross_validation import cross_val_score
# loop over the different levels of max_depth, calculate 10-fold cross validation MSE for each
for x in max_depth:
    regressor = DecisionTreeRegressor(max_depth=x, random_state=0)
    RMSE_scores.append(np.mean(np.sqrt(-(cross_val_score(regressor, X=dummies1[feature_cols], y=charged_dummies, cv=10, scoring='mean_squared_error'))))) 
```

```python
plt.plot(max_depth, RMSE_scores)
plt.xlabel('max_depth')
plt.ylabel('RMSE (lower is better)')
```

```python
sorted(zip(RMSE_scores, max_depth))[0]
```

```python
treereg = DecisionTreeRegressor(max_depth=2)
treereg.fit(X, y)
```

```python
pd.DataFrame({'feature':feature_cols, 'importance':treereg.feature_importances_}).sort('importance')
```

```python
treereg = DecisionTreeRegressor(max_depth=3)
treereg.fit(X, y)
```

```python
pd.DataFrame({'feature':feature_cols, 'importance':treereg.feature_importances_}).sort('importance')
```

```python
from sklearn.ensemble import RandomForestRegressor
estimator_range = range(10, 310, 10)
RMSE_scores = []
for estimator in estimator_range:
    rfreg=RandomForestRegressor(n_estimators=estimator,random_state=1)
    MSE_scores=cross_val_score(rfreg,X,y,cv=5,scoring='mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
```

```python
# plot n_estimators (x-axis) versus RMSE (y-axis)
plt.plot(estimator_range, RMSE_scores)
plt.xlabel('n_estimators')
plt.ylabel('RMSE (lower is better)')
```

```python
sorted(zip(RMSE_scores, estimator_range))[0]
```

```python
feature_range = range(1, len(feature_cols)+1)
RMSE_scores = []
for feature in feature_range:
    rfreg=RandomForestRegressor(max_features=feature,random_state=1)
    MSE_scores=cross_val_score(rfreg,X,y,cv=10,scoring='mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))
```

```python
plt.plot(feature_range, RMSE_scores)
plt.xlabel('max_features')
plt.ylabel('RMSE (lower is better)')
```

```python
sorted(zip(RMSE_scores, feature_range))[0]
```

```python
max_depth = range(1,11)

# first we create a  list to store the average RMSE for each value of max_depth
RMSE_scores = []

from sklearn.cross_validation import cross_val_score
# loop over the different levels of max_depth, calculate 10-fold cross validation MSE for each
for x in max_depth:
    regressor = DecisionTreeRegressor(max_depth=x, random_state=0)
    RMSE_scores.append(np.mean(np.sqrt(-(cross_val_score(regressor, X=dummiesdispX[feature_cols], y=disp_dummies['disp_Justified'], cv=10, scoring='mean_squared_error')))))
```

```python
plt.plot(max_depth, RMSE_scores)
plt.xlabel('max_depth')
plt.ylabel('RMSE (lower is better)')
```

```python
from sklearn.ensemble import RandomForestRegressor
estimator_range = range(10, 310, 10)
RMSE_scores = []
for estimator in estimator_range:
    rfreg=RandomForestRegressor(n_estimators=estimator,random_state=1)
    MSE_scores=cross_val_score(rfreg,X,y,cv=5,scoring='mean_squared_error')
    RMSE_scores.append(np.mean(np.sqrt(-MSE_scores)))

plt.plot(estimator_range, RMSE_scores)
plt.xlabel('n_estimators')
plt.ylabel('RMSE (lower is better)')
```

```python
sorted(zip(RMSE_scores, estimator_range))[0]
```
